## LIST

```python
l1 = [1,23,45,67]
l2 =[34,49,23,64]
import numpy as np
a1 = np.array(l1)
a2 = np.array(l2)
print(a1)
print(a2)
print(type(a1))
print(type(a2))
print("Addition of List 1 and List 2:")
print(a1+a2)
print("Subtraction of List 1 and List 2:")
print(a1-a2)
print("Multiplication of List 1 and List 2:")
print(a1*a2)
print("Division of List 1 and List 2:")
print(a1/a2)
print("Modulus of List 1 and List 2:")
print(a1%a2)
print("Exponent of List 1 and List 2:")
print(a1**a2)
print("Floor Division of List 1 and List 2:")
print(a1//a2)
print("sin of list 1")
print(np.sin(a1))
print("cos of list 1")
print(np.cos(a1))
print("tan of list 1")
print(np.tan(a1))
print("log of list 1")
print(np.log(a1))
print("log10 of list 1")
print(np.exp(a1))
print("log2 of list 1")
print(np.log2(a1))
print(np.power(a1,a2))
```

## ARRAY

## Part 2 – Visualization of Data using matplotlib, pyplots Packages[CO2]

```python
#Generate a Data set for Health Care using the following:
people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]
import matplotlib.pyplot as plt
plt.plot(age,height)
plt.xlabel("Age")
plt.ylabel("Height")
plt.title("Age vs Height")
plt.show()
```

```python
plt.plot(age,weight)
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
#2. Using the generated dataset given above. Generate the following graphs and justify the relationships among the vectors.
#Scatter plot
plt.scatter(age,height)
plt.xlabel("Age")
plt.ylabel("Height")
plt.title("Age vs Height")
plt.show()
```

```python
plt.scatter(age,weight)
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
#Bar Chart
plt.bar(people,age)
plt.xlabel("People")
plt.ylabel("Age")
plt.title("People vs Age")
plt.show()
```

```python
plt.bar(people,weight)
plt.xlabel("People")
plt.ylabel("Weight")
plt.title("People vs Weight")
plt.show()
```

```python
#5. Histogram
plt.hist(age)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.show()
```

```python
plt.hist(weight)
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.title("Histogram of Weight")
plt.show()
```
